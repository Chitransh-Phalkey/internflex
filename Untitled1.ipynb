```python
# Sabse pehle Deep Learning aur plots ke zaroori tools import kar rahe hain
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, GlobalAveragePooling1D, Dense
import matplotlib.pyplot as plt
import numpy as np

# IMDb dataset load kar rahe hain (Fix bracket unpacking)
num_words = 10000
(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=num_words)

print("Data successfully load ho gaya!")
print(f"Training reviews: {len(X_train)}, Testing reviews: {len(X_test)}")
```

```python
# Sabhi reviews ki maximum length 250 words set kar rahe hain
max_len = 250

# pad_sequences ka use karke sabhi reviews ko ek barabar size ka bana rahe hain
X_train_padded = pad_sequences(X_train, maxlen=max_len, padding='post', truncating='post')
X_test_padded = pad_sequences(X_test, maxlen=max_len, padding='post', truncating='post')

print("Padding complete ho gayi!")
print("Pehle review ka naya shape:", X_train_padded[0].shape)
```

```python
# Sabhi reviews ki maximum length 250 words set kar rahe hain
max_len = 250

# pad_sequences ka use karke sabhi reviews ko ek barabar size ka bana rahe hain
X_train_padded = pad_sequences(X_train, maxlen=max_len, padding='post', truncating='post')
X_test_padded = pad_sequences(X_test, maxlen=max_len, padding='post', truncating='post')

print("Padding complete ho gayi!")
print("Pehle review ka naya shape:", X_train_padded[0].shape)
```

```python
# Sequential model initialize kar rahe hain
model = Sequential([
    Embedding(input_dim=10000, output_dim=16, input_length=250),
    GlobalAveragePooling1D(),
    Dense(16, activation='relu'),
    Dense(1, activation='sigmoid') # Binary Output: 0 ya 1
])

# Model ko compile kar rahe hain (Optimizer aur Loss function set kar rahe hain)
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Model ka structure dekhne ke liye summary print kar rahe hain
model.summary()
```

```python
# Sequential model initialize kar rahe hain
model = Sequential([
    Embedding(input_dim=10000, output_dim=16, input_length=250),
    GlobalAveragePooling1D(),
    Dense(16, activation='relu'),
    Dense(1, activation='sigmoid') # Binary Output: 0 ya 1
])

# Model ko compile kar rahe hain (Optimizer aur Loss function set kar rahe hain)
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Model ka structure dekhne ke liye summary print kar rahe hain
model.summary()
```

```python
print("Model ki training shuru ho rahi hai aur iske baad direct graph banega...\n")

# 1. Model ko fir se train kar rahe hain taaki history variable ban jaye
history = model.fit(
    X_train_padded, y_train,
    epochs=10,
    batch_size=128,
    validation_data=(X_test_padded, y_test),
    verbose=1
)

print("\nTraining complete! Ab graphs generate ho rahe hain...\n")

# 2. Graphs banane ka code
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs_range = range(1, len(acc) + 1)

plt.figure(figsize=(14, 5))

# Accuracy Plot
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy', color='blue', marker='o')
plt.plot(epochs_range, val_acc, label='Validation Accuracy', color='orange', marker='o')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')
plt.grid(True)

# Loss Plot
plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss', color='blue', marker='o')
plt.plot(epochs_range, val_loss, label='Validation Loss', color='orange', marker='o')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend(loc='upper right')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# 1. Model ko save kar rahe hain (Deliverable Requirement)
model.save('sentiment_model.h5')
print("Model successfully save ho gaya: 'sentiment_model.h5' file ban gayi hai!\n")

# 2. Inference Function (Naye text par predict karne ke liye)
def predict_sentiment(custom_review):
    # IMDb ke rules ke hisab se text ko simple words mein split kar rahe hain
    word_index = imdb.get_word_index()
    words = custom_review.lower().split()

    # Words ko unke numerical tokens mein convert kar rahe hain
    tokenized = []
    for word in words:
        if word in word_index and word_index[word] < 10000:
            tokenized.append(word_index[word])
        else:
            tokenized.append(2) # 2 ka matlab unknown word

    # Padding kar rahe hain taaki length 250 ho jaye
    padded = pad_sequences([tokenized], maxlen=250, padding='post', truncating='post')

    # Model se prediction le rahe hain
    prediction = model.predict(padded)[0][0]

    print(f"\nReview: '{custom_review}'")
    print(f"Prediction Score: {prediction:.4f}")
    if prediction >= 0.5:
        print("Result: 😍 POSITIVE REVIEW!")
    else:
        print("Result: 😡 NEGATIVE REVIEW!")

# 3. Test karke dekhte hain!
predict_sentiment("this movie was absolutely amazing and wonderful")
predict_sentiment("the plot was terrible and waste of time")
```
