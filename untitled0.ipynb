```python
# Sabse pehle zaroori tools (libraries) import kar rahe hain
import pandas as pd
import numpy as np

# Internet se direct standard dataset load kar rahe hain
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/iris.csv"
names = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'class']
dataset = pd.read_csv(url, names=names)

# Dekhte hain hamara data kaisa dikhta hai (Pehli 5 rows)
print("--- Data ki pehli 5 rows ---")
print(dataset.head())
```

```python
# 1. Check karte hain data ka shape (Rows, Columns)
print("Dataset ka size (Rows, Columns):", dataset.shape)

# 2. Check karte hain ki koi value missing (khali) to nahi hai
print("\nMissing values check:")
print(dataset.isnull().sum())
```

```python
# train_test_split tool ko import kar rahe hain
from sklearn.model_selection import train_test_split

# X mein saari columns daal rahe hain 'class' ko chhod kar (Inputs)
X = dataset.drop('class', axis=1)

# y mein sirf 'class' waali column daal rahe hain (Target/Output)
y = dataset['class']

# Data ko 80% Train aur 20% Test mein split kar rahe hain
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=1, stratify=y)

# Check karte hain ki split sahi se hua ya nahi
print("Training data ka size (80%):", X_train.shape)
print("Testing data ka size (20%):", X_test.shape)
```

```python
# Dono algorithms ko import kar rahe hain
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

# 1. Logistic Regression model ko train kar rahe hain
lr_model = LogisticRegression(max_iter=200)
lr_model.fit(X_train, y_train)

# 2. Random Forest model ko train kar rahe hain
rf_model = RandomForestClassifier(random_state=1)
rf_model.fit(X_train, y_train)

print("Dono models successfully train ho gaye hain!")
```

```python
# Dono models se test data par predictions karwa rahe hain
lr_predictions = lr_model.predict(X_test)
rf_predictions = rf_model.predict(X_test)

# Dekhte hain dono ke kuch predictions kaise hain
print("Asli phool kaunse the (Actual):")
print(list(y_test)[:5])

print("\nLogistic Regression ne kya guess kiya:")
print(list(lr_predictions)[:5])

print("\nRandom Forest ne kya guess kiya:")
print(list(rf_predictions)[:5])
```

```python
# Metrics aur Plotting ke tools import kar rahe hain
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# 1. LOGISTIC REGRESSION KA REPORT CARD
print("=== LOGISTIC REGRESSION METRICS ===")
print(f"Accuracy : {accuracy_score(y_test, lr_predictions):.2f}")
print(f"Precision: {precision_score(y_test, lr_predictions, average='macro'):.2f}")
print(f"Recall   : {recall_score(y_test, lr_predictions, average='macro'):.2f}")
print(f"F1-Score : {f1_score(y_test, lr_predictions, average='macro'):.2f}")

# 2. RANDOM FOREST KA REPORT CARD
print("\n=== RANDOM FOREST METRICS ===")
print(f"Accuracy : {accuracy_score(y_test, rf_predictions):.2f}")
print(f"Precision: {precision_score(y_test, rf_predictions, average='macro'):.2f}")
print(f"Recall   : {recall_score(y_test, rf_predictions, average='macro'):.2f}")
print(f"F1-Score : {f1_score(y_test, rf_predictions, average='macro'):.2f}")

# 3. CONFUSION MATRIX PLOT (GRAPH)
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

ConfusionMatrixDisplay.from_predictions(y_test, lr_predictions, ax=ax[0], cmap='Blues')
ax[0].set_title("Logistic Regression")

ConfusionMatrixDisplay.from_predictions(y_test, rf_predictions, ax=ax[1], cmap='Greens')
ax[1].set_title("Random Forest")

plt.tight_layout()
plt.show()
=
```

```python
# cross_val_score tool import kar rahe hain
from sklearn.model_selection import cross_val_score

# Logistic Regression par 5-Fold Cross Validation check kar rahe hain
lr_cv = cross_val_score(lr_model, X, y, cv=5)
print("Logistic Regression CV Scores (5 baar ka test):", lr_cv)
print(f"Mean CV Accuracy: {lr_cv.mean():.2f}")

# Random Forest par 5-Fold Cross Validation check kar rahe hain
rf_cv = cross_val_score(rf_model, X, y, cv=5)
print("\nRandom Forest CV Scores (5 baar ka test):", rf_cv)
print(f"Mean CV Accuracy: {rf_cv.mean():.2f}")
```
